```python
user_segement = pd.read_csv("/Users/Ryu Seung Gwon/sgryu/대외활동/코멘토/sql/week4/_VIP_유저_국가별_분포_SELECT_FROM_WITH_RecencyTarget_AS_SELECT_WID_MAX__202402020036.csv")
```

```python
user_segement = user_segement[['Segment', 'country']]
```

```python
user_segement.dropna(inplace=True)
```

```python
user_segement['Segment'].value_counts()
```

```python
vip = user_segement[user_segement['Segment'] == 'VIP']
smallpay = user_segement[user_segement['Segment'] == '소과금 VIP']
midpay = user_segement[user_segement['Segment'] == '중과금 VIP']
leavevip = user_segement[user_segement['Segment'] == '이탈 VIP']
potenitalvip = user_segement[user_segement['Segment'] == '잠재 VIP']
```

```python
country_counts = smallpay['country'].value_counts()
rows_to_remove = smallpay[smallpay['country'].map(country_counts) == 1].index
smallpay = smallpay.drop(rows_to_remove)
smallpay
```

```python
sns.set(style='darkgrid')
font_path = "/Library/Fonts/Arial Unicode.ttf"
font_name = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

plt.figure(figsize=(10, 8))
ax = sns.countplot(x='Segment', hue='country', data=vip, palette="Set2")

# 각 막대 위에 빈도를 표시
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 10),
                textcoords='offset points')

plt.xlabel('국가', fontsize=15)
plt.ylabel('빈도', fontsize=15)
plt.title("국가 별 VIP 유저 수", fontsize=18)
plt.ticklabel_format(style='plain', axis='y')
plt.show()
```
